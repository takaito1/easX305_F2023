# Reading GO-SHIP bottle data

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Open the cruise data file
#
# define file name
fn='33RO20161119_hy1.csv'
# read in the cruise data
df=pd.read_csv(fn,skiprows=149,header=0)
# display the first 5 rows
df.head()
```

```python
# display the name of the all columns
df.columns
```

```python
# identify station #9
stn9 = (df['STNNBR']==9)
# extract the data belonging to station #9 and name it as df9
df9 = df[stn9]
df9.head()
```

```python
# Get all the T and S data points for Station 9
T = df9['CTDTMP'].to_numpy().astype(float)
S = df9['CTDSAL'].to_numpy().astype(float)
```
